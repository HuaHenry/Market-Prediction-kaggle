# Basic Training Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (8, 5)
plt.rcParams["axes.grid"] = True

# 设置dpi
plt.rcParams["figure.dpi"] = 300

df = pd.read_csv("../data/train.csv")
print(df.shape)
df.head()
```

缺失值情况 & 数值特征基本统计

```python
# 缺失值整体情况
missing = df.isna().sum()
missing_pct = missing / len(df) * 100
missing_df = (
    pd.DataFrame({"missing_count": missing, "missing_pct": missing_pct})
    .sort_values("missing_pct", ascending=False)
)

print(missing_df.head(30))

# 画缺失率最高的前20个特征柱状图
top_missing = missing_df.head(20)
plt.figure(figsize=(10, 4))
plt.bar(top_missing.index.astype(str), top_missing["missing_pct"])
plt.xticks(rotation=90)
plt.ylabel("Missing (%)")
plt.title("Top 20 Features by Missing Percentage")
plt.tight_layout()
plt.show()

# 数值特征的描述统计
num_desc = df.describe(include=[np.number]).T
num_desc.head(30)
```

识别时间列

```python
# 自动尝试寻找日期/时间列
date_col = None
for col in df.columns:
    if any(k in col.lower() for k in ["date", "time", "timestamp"]):
        try_series = pd.to_datetime(df[col], errors="coerce", infer_datetime_format=True)
        if try_series.notna().mean() > 0.5:
            df[col] = try_series
            date_col = col
            break

print("date_col:", date_col)
```

三个关键变量：统计特性 + 相关矩阵

```python
key_cols = ["forward_returns", "risk_free_rate", "market_forward_excess_returns"]
available_key_cols = [c for c in key_cols if c in df.columns]

df_key = df[available_key_cols].copy()

# 统计特性
key_summary = df_key.describe().T
print(key_summary)

# 相关矩阵
key_corr = df_key.corr()
print(key_corr)

# 可视化相关矩阵（简单 heatmap）
plt.figure(figsize=(5, 4))
im = plt.imshow(key_corr, aspect="auto")
plt.xticks(range(len(available_key_cols)), available_key_cols, rotation=45, ha="right")
plt.yticks(range(len(available_key_cols)), available_key_cols)
plt.colorbar(im)
plt.title("Correlation Matrix of Key Variables")
plt.tight_layout()
plt.show()
```

forward_returns 与 market_forward_excess_returns 的散点关系

```python
if {"forward_returns", "market_forward_excess_returns"}.issubset(df.columns):
    sample = df[["forward_returns", "market_forward_excess_returns"]].dropna()
    if len(sample) > 5000:
        sample = sample.sample(5000, random_state=42)

    plt.figure(figsize=(6, 5))
    plt.scatter(
        sample["forward_returns"],
        sample["market_forward_excess_returns"],
        alpha=0.3,
        marker="x"
    )
    plt.xlabel("forward_returns")
    plt.ylabel("market_forward_excess_returns")
    plt.title("forward_returns vs. market_forward_excess_returns")
    plt.tight_layout()
    plt.show()
```

risk_free_rate 与 forward_returns 的散点关系

```python
if {"forward_returns", "risk_free_rate"}.issubset(df.columns):
    sample = df[["risk_free_rate", "forward_returns"]].dropna()
    if len(sample) > 5000:
        sample = sample.sample(5000, random_state=42)

    plt.figure(figsize=(6, 5))
    plt.scatter(
        sample["risk_free_rate"],
        sample["forward_returns"],
        alpha=0.3,
        marker="x"
    )
    plt.xlabel("risk_free_rate")
    plt.ylabel("forward_returns")
    plt.title("risk_free_rate vs. forward_returns")
    plt.tight_layout()
    plt.show()
```

构造 fr_minus_rf，并与 market_forward_excess_returns 比较

```python
if {"forward_returns", "risk_free_rate", "market_forward_excess_returns"}.issubset(df.columns):
    df_rel = df[["forward_returns", "risk_free_rate", "market_forward_excess_returns"]].dropna().copy()
    df_rel["fr_minus_rf"] = df_rel["forward_returns"] - df_rel["risk_free_rate"]
    df_rel["diff_excess_vs_market"] = df_rel["fr_minus_rf"] - df_rel["market_forward_excess_returns"]

    # 统计表
    rel_desc = df_rel[["fr_minus_rf", "market_forward_excess_returns", "diff_excess_vs_market"]].describe().T
    print(rel_desc)

    # 差值直方图
    plt.figure(figsize=(6, 4))
    df_rel["diff_excess_vs_market"].hist(bins=50)
    plt.xlabel("(forward_returns - risk_free_rate) - market_forward_excess_returns")
    plt.ylabel("Frequency")
    plt.title("Difference Distribution")
    plt.tight_layout()
    plt.show()
```

三个变量的分布直方图

```python
for col in available_key_cols:
    plt.figure(figsize=(6, 4))
    df[col].dropna().hist(bins=50)
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.title(f"Distribution of {col}")
    plt.tight_layout()
    plt.show()
```

目标变量随时间变化

```python
target_col = "market_forward_excess_returns"  # 如果你想换成 forward_returns 就改这里

if date_col is not None and target_col in df.columns:
    tmp = df[[date_col, target_col]].dropna().sort_values(date_col)
    # 只画前 5000 点，避免太密集
    tmp = tmp.iloc[:5000]

    plt.figure(figsize=(10, 4))
    plt.plot(tmp[date_col], tmp[target_col])
    plt.xlabel(date_col)
    plt.ylabel(target_col)
    plt.title(f"{target_col} Over Time")
    plt.tight_layout()
    plt.show()
```
